## Multi-Hot Label Encoding

* The NIH dataset stores labels as pipe-separated strings like `"Atelectasis|Effusion"`.
* Convert these into binary vectors of length 14 (one entry per disease class).
* `No Finding` is treated as all-zeros: the model doesn't need a 15th output for it.

```python
CLASSES = [
    'Atelectasis', 'Cardiomegaly', 'Consolidation', 'Edema', 'Effusion', 'Emphysema', 'Fibrosis', 'Hernia', 'Infiltration', 'Mass', 'Nodule', 'Pleural_Thickening', 'Pneumonia', 'Pneumothorax'
    ]

NUM_CLASSES = len(CLASSES)
print(f'Number of classes: {NUM_CLASSES}')
print(f'Classes: {CLASSES}')
```

```python
def encode_labels(label_string: str, classes:list) -> np.ndarray:
    """
    Convert a pipe-separated label string to a multi-hot binary vector

    Args:
        label string: e.g. 'Atelectasis|Effusion' or 'No Finding'
        classes: ordered list of 14 disease class names

    Returns:
        np.ndarray of shape (14,) with 0s and 1s
    """
    vec = np.zeros(len(classes), dtype=np.float32)
    if label_string == 'No Finding':
        return vec # all zeros
    for label in label_string.split('|'):
        label = label.strip() # removes whitespace form the beginning and end of the string
        if label in classes:
            vec[classes.index(label)] = 1.0
    return vec
```

```python
label_string_sample = df.iloc[1]['Finding Labels']
label_string_sample
```

```python
label_matrix = np.stack(
    df['Finding Labels'].apply(lambda x: encode_labels(x, CLASSES)).values
)

print(label_matrix.shape)

# Add multi-hot labels to the dataframe
df['labels'] = list(label_matrix)
```

Random splits leak patient data across train/test because the same patient can have multiple X-rays. The official split is patient-level.

We carve 10% of the official `train_val_list` set for validation (still patient-aware since the NIH file already handles patient-level separation).

* `Validation set`: Used for tuning and model selection.
* `Test set`: Used for final evaluation on unseen data.

# Pytorch Dataset and DataLoader

```python
class NIHChestXrayDataset(Dataset):
    """
    PyTorch Dataset for NIH Chest X-ray.

    Args:
        dataframe: pandas Dataframe with 'Image Index' and 'labels columns'
        filename_to_path: dict mapping filename -> full Path object
        transform: torchvision transform pipeline

    Returns:
        tensor of image and labels[shape (14) with 0s and 1s]
    """


    def __init__(self, dataframe: pd.DataFrame, filename_to_path: dict, transform=None):
        self.df = dataframe.reset_index(drop=True)
        self.filename_to_path = filename_to_path
        self.transform = transform

    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, index):
        row = self.df.iloc[index]
        img_path = self.filename_to_path[row['Image Index']]

        # NIH images are grayscale PNGs - so we need to convert them to RGB so DenseNet-121 can accept them (DenseNet expects 3-channel input)
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)
        
        labels = torch.tensor(row['labels'], dtype=torch.float32)

        return image, labels
```

```python
# Densenet: All pre-trained models expect input images normalized in the same way, i.e. mini-batches of 3-channel RGB images of shape (3 x H x W), where H and W are expected to be at least 224. The images have to be loaded in to a range of [0, 1] and then normalized using mean = [0.485, 0.456, 0.406] and std = [0.229, 0.224, 0.225].

# ImageNet normalization stats (this is required because DenseNet-121 was pretrained on ImageNet)
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD = [0.229, 0.224, 0.225]


train_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD)
])

val_test_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD)
])


# Instantiate datasets
train_dataset = NIHChestXrayDataset(df_train, filename_to_path, transform=train_transform)

val_dataset = NIHChestXrayDataset(df_val, filename_to_path, transform=val_test_transform)

test_dataset = NIHChestXrayDataset(df_test, filename_to_path, transform=val_test_transform)

print(f'Train dataset: {len(train_dataset):,} samples')
print(f'Val dataset: {len(val_dataset):,} samples')
print(f'Test dataset: {len(test_dataset):,} samples')
```

```python
BATCH_SIZE = 16
NUM_WORKERS = 0

train_dataloader = DataLoader(
    dataset=train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    pin_memory=True if device.type == 'cuda' else False,
    drop_last=True, # avoids incomplete batch issues with BatchNorm
)

val_dataloader = DataLoader(
    dataset=val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True if device.type == 'cuda' else False,
)

test_dataloader = DataLoader(
    dataset=test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True if device.type == 'cuda' else False,
)

print(f'Train batches: {len(train_dataloader):,}')
print(f'Val batches: {len(val_dataloader):,}')
print(f'Test batches: {len(test_dataloader):,}')
```

```python
# Verify one batch loads correctly

images, labels = next(iter(train_dataloader))

print(f'Image batch shape: {images.shape}')     # expect (BATCH_SIZE, 3, 224, 224)
print(f'Labels batch shape: {labels.shape}')    # expect (BATCH_SIZE, 14)
print(f'Image dtype: {images.dtype}')           # float32
print(f'Labels dtype : {labels.dtype}')         # float32
print(f'Image value range : [{images.min():.3f}, {images.max():.3f}]')  # normalized values

print(f'\nSample label vectors (first 6 in batch):')
for i in range(6):
    active = [CLASSES[j] for j, v in enumerate(labels[i]) if v == 1.0]
    print(f'    Sample {i+1}: {active if active else "[No Finding]"}')
```

```python
# Visualize a few images form the batch to confirm pipeline is right

def denormalize(tensor, mean=IMAGENET_MEAN, std=IMAGENET_STD):
    """
    Reverse ImageNet normalization for display.
    """
    t = tensor.clone()
    for c, (m, s) in enumerate(zip(mean, std)):
        t[c] = t[c] * s + m
    return t.clamp(0, 1)

fig, axes = plt.subplots(2, 4, figsize=(14, 7))
axes = axes.flatten()

for i in range(8):
    img_display = denormalize(images[i]).permute(1, 2, 0).numpy()
    axes[i].imshow(img_display, cmap='gray')
    active = [CLASSES[j] for j, v in enumerate(labels[i]) if v == 1.0]
    title = '\n'.join(active) if active else 'No Finding'
    color = 'red' if active else 'green'
    axes[i].set_title(title, fontsize=8, color=color)
    axes[i].axis('off')


plt.suptitle('Sample Batch from DataLoader (after transforms)', fontsize=12)
plt.tight_layout()
plt.show()
```

# LOAD DenseNet-121

```python
def build_densenet121(num_classes: int = 14, pretrained: bool = True) -> nn.Module:
    """
    Load pretrained DenseNet-121 and replace the classifier head to output `num_classes` logits for multi-label classification.
    """

    model = models.densenet121(weights='IMAGENET1K_V1' if pretrained else None)

    # DenseNet-121 classifier: Linear(1024 features -> 1000 logits output)
    # We will replace with: Linear(1024 features -> num_classes)

    in_features = model.classifier.in_features # 1024 for DenseNet-121
    model.classifier = nn.Linear(in_features, num_classes)

    # Xavier (Glorot) Uniform init for the new head

    # The new head starts with randomly initialized weights. Xavier initialization
    # Glorot chooses the weight range based on the number of input and output units
    # this helps keep activation and gradient magnitudes reasonably stable during training (solves vanishing or exploding gradients by keeping the variance of activations and gradients roughly consistent across network layers, ensuring stable learning).

    nn.init.xavier_uniform_(model.classifier.weight)
    nn.init.zeros_(model.classifier.bias)

    return model
```

```python
model = build_densenet121(num_classes=NUM_CLASSES, pretrained=True)
model = model.to(device)

print(f'Model: DenseNet-121')
print(f'Classifier head: {model.classifier}')
print(f'Device: {next(model.parameters()).device}')

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f'\nTotal Parameters: {total_params:,}')
print(f'Trainable Parameters: {trainable_params:,}')
```

```python
# Forward pass of one batch into the loaded DenseNet-121 Model

model.eval()

images, labels = next(iter(train_dataloader))

with torch.no_grad():
    sample_images = images.to(device)
    logits = model(sample_images)
    probs = torch.sigmoid(logits) # sigmoid activation funtion maps any y(input) value in range of (0, 1)

print('Forward Pass')
print(f'Input shape: {sample_images.shape}')
print(f'Output shape: {logits.shape}')
print(f'Output dtype: {logits.dtype}')
print(f'Logit range: [{logits.min().item():.4f}, {logits.max().item():.4f}]')
print(f'Prob range: [{probs.min().item():.4f}, {probs.max().item():.4f}]') # in (0, 1)
print(f'\nSample output - first image, per-class probabilities:')

for cls, prob in zip(CLASSES, probs[0].cpu().numpy()):
    print(f'    {cls:25s}: {prob:.4f}')
```

## Loss Function + Class Weights
